```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import math
import re
```

```python
f_sensor=open('weallwalk/T1_ID1_1L_GD_iPhoneSensors.csv')  
dataset=pd.read_csv(f_sensor)
#dataset.head(10)

data_x = np.array(dataset.iloc[:, 0:39])
print(data_x.shape)
print(data_x)

data_y = np.array(dataset.iloc[:, 40:41])
data_y = np.array(data_y)
    
print data_y.shape
print data_y

import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d  

plt.figure(figsize=(18, 5))
plt.plot([i[0] for i in data_x], [i[0] for i in data_y], 'y')
plt.show()
```

```python
f = open('weallwalk/T1_ID1_GD.xml')
cts = f.read()
f.close()

print cts[0:100]

p_foot = re.compile(r'<WhichFoot>(.*?)</WhichFoot>')
all_foot = p_foot.findall(cts)
print len(all_foot)

p_time = re.compile(r'<Time>(.*?)</Time>')
all_time = p_time.findall(cts)
print len(all_time)

strike_times = []
for i in range(len(all_foot)):
    strike_times.append([float(all_time[i]), all_foot[i]])
#     print all_foot[i]
#     print float(all_time[i])

print len(strike_times)

window_y = []
# left - 0, right = 1
strike_times.append([len(data_x)/25.0, 'E'])
step_direct = 0.0
strike_index=0
for i in range(len(data_x)):
    if(strike_index<len(strike_times) and i/25.0 < strike_times[strike_index][0]):
        window_y.append(step_direct)
    else:
        step_direct = 1-step_direct
        strike_index += 1
        window_y.append(step_direct)

plt.figure(figsize=(18, 5))
plt.plot([i[0] for i in data_x], [i for i in data_y], 'y')
plt.plot([i[0] for i in data_x], [i for i in window_y], 'b')
plt.show()
```

```python
from tensorflow.contrib import rnn

# matplotlibsin
from tensorflow.contrib.learn.python.learn.estimators.estimator import SKCompat
from matplotlib import pyplot as plt

learn = tf.contrib.learn
HIDDEN_SIZE = 50  # Lstm
NUM_LAYERS = 2  # LSTM
TIMESTEPS = 100  # 
TRAINING_STEPS = 10000  # 
BATCH_SIZE = 32  # batch
```

```python
train_x,train_y=[],[] 
for i in range(0, 1300):
#     x = [[d[5] for d in data_x[i:i+TIMESTEPS]]]
    x = data_x[i:i+TIMESTEPS]
    x = np.transpose(x)
    y= [window_y[i+TIMESTEPS+1]]
    train_x.append(x)
    train_y.append(y)
    
test_x, test_y=[],[] 
for i in range(1300, 2100):
#     x = [[d[5] for d in data_x[i:i+TIMESTEPS]]]
    x = data_x[i:i+TIMESTEPS]
    x = np.transpose(x)
    y= [window_y[i+TIMESTEPS+1]]
    test_x.append(x)
    test_y.append(y)
    
train_x = np.array(train_x, dtype=np.float32)
train_y = np.array(train_y, dtype=np.float32)
    
test_x = np.array(test_x, dtype=np.float32)
test_y = np.array(test_y, dtype=np.float32)

print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)

# (1200, 1, 100)
# (1200, 1)
# (800, 1, 100)
# (800, 1)
```

```python
def LstmCell():
    lstm_cell = rnn.BasicLSTMCell(HIDDEN_SIZE,state_is_tuple=True)
    return lstm_cell

# lstm
def lstm_model(X, y):
    print X.shape
    cell = rnn.MultiRNNCell([LstmCell() for _ in range(NUM_LAYERS)])
    output, _ = tf.nn.dynamic_rnn(cell, X, dtype=tf.float32)
    output = tf.reshape(output, [-1, HIDDEN_SIZE])
    print output.shape
    # 
    predictions = tf.contrib.layers.fully_connected(output, 1, None)
    
    # predictionslabelsshape
    labels = tf.reshape(y, [-1])
    predictions = tf.reshape(predictions, [-1])
    
    print labels.shape
    print predictions.shape
    loss = tf.losses.mean_squared_error(predictions, labels)
    train_op = tf.contrib.layers.optimize_loss(loss, tf.train.get_global_step(),
                                             optimizer="Adagrad",
                                             learning_rate=0.5)
    return predictions, loss, train_op

# 
# lstm
regressor = SKCompat(learn.Estimator(model_fn=lstm_model))
# 
regressor.fit(train_x, train_y, batch_size=BATCH_SIZE, steps=TRAINING_STEPS)
```

```python
predicted = [[pred] for pred in regressor.predict(train_x)]

# MSE
rmse = np.sqrt(((predicted - train_y) ** 2).mean(axis=0))
print("Mean Square Error is:%f" % rmse[0])

plt.figure(figsize=(18, 5))
plot_predicted, = plt.plot(predicted, label='predicted')
plot_test, = plt.plot(train_y, label='train_y')
plt.legend([plot_predicted, plot_test],['predicted', 'train_y'])
plt.show()
```

```python
predicted = [[pred] for pred in regressor.predict(test_x)]

# MSE
rmse = np.sqrt(((predicted - test_y) ** 2).mean(axis=0))
print("Mean Square Error is:%f" % rmse[0])

plt.figure(figsize=(18, 5))
plot_predicted, = plt.plot(predicted, label='predicted')
plot_test, = plt.plot(test_y, label='test_y')
plt.legend([plot_predicted, plot_test],['predicted', 'test_y'])
plt.show()
```
